```python
from mod3d import( 
    BRepBuilderAPI as brb,
    GeomFill as gfi,
    BRepOffsetAPI as bro,
    gp,   
)

c1 = gp.Circ(gp.Ax2(gp.Pnt(0, 0, 0), gp.Dir(0,0,1)), 0.4)

profile1      = brb.MakeWire( brb.MakeEdge(c1).edge() ).wire()
profile1_face = brb.MakeFace( profile1 ).face()

e1 = brb.MakeEdge(gp.Pnt(0, 0, 0), gp.Pnt(10, 0, 0)).edge()
e2 = brb.MakeEdge(gp.Pnt(10, 0, 0), gp.Pnt(15, 5, 0)).edge()


guide = brb.MakeWire(e1, e2).wire()

poly = brb.MakePolygon()
poly.add(gp.Pnt(0,0,0))
poly.add(gp.Pnt(10,0,0))
poly.add(gp.Pnt(15,5,0))
poly.add(gp.Pnt(15,10,0))
poly.add(gp.Pnt(20,10,0))
guide = poly.wire()

# builder = brb.MakePipe(guide, profile, gfi., force_approx_c1=True)
# endo = builder.shape()

```

```python
# builder.set_transition_mode(brb.TransitionMode.Transformed)
# builder.set_discrete_mode()
# builder.set_transition_mode(brb.TransitionMode.RoundCorner)

builder = brb.MakePipeShell(guide)
builder.add(profile1, with_contact= False, with_correction=True)
builder.set_transition_mode(brb.TransitionMode.RightCorner)
# builder.set_transition_mode(brb.TransitionMode.RoundCorner)
# builder.set_force_approx_c1(True)
print(builder.get_status())
builder.build()
endo = builder.shape()
```

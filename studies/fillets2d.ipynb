```python
from mod3d import (
    gp,
    BRepBuilderAPI,
    BRepFillet,
    TopExp,
    TopAbs,
    TopoDS,
)


# Create a rectangular face
p1 = gp.Pnt(0.0, 0.0, 0.0)
p2 = gp.Pnt(10.0, 0.0, 0.0)
p3 = gp.Pnt(10.0, 5.0, 0.0)
# p4 = gp.Pnt(0.0, 5.0, 0.0)

# Use MakePolygon for proper vertex merging
wire_maker = BRepBuilderAPI.MakePolygon(p1, p2, p3, False)
wire = wire_maker.wire()
face = BRepBuilderAPI.MakeFace(wire).face()

# Get vertices and cast them to Vertex
vertices = [TopoDS.Vertex(v) for v in TopExp.Explorer(wire, TopAbs.ShapeEnum.VERTEX)]

# Create fillet2d builder
fillet2d = BRepFillet.MakeFillet2d(face)

# Add fillet at first vertex (corner)
fillet_edge = fillet2d.add_fillet(vertices[1], 1.0)

fillet2d.build()
```

```python
from mod3d import StepControl as stc

writer = stc.Writer()
status = writer.transfer(fillet2d.shape(), stc.AsIs)

write_status = writer.write("filet2d.step") 
```

```python
from mod3d import BRepBuilderAPI as brb
from mod3d import gp
import math

corps_moteur = brb.MakeCylinder(r = 10, h = 50).solid()

pos0 = gp.Trsf()
pos0.set_translation(gp.Vec(0, 0, -25))
corps_moteur.move(pos0)

box1 = brb.MakeBox(dx=1, dy=1, dz=3).solid()
pos1 = gp.Trsf()
pos1.set_translation(gp.Vec(10.51, 0, 0))
box1.move(pos1)

box2 = brb.MakeBox(dx=2, dy=1, dz=3).solid()
pos2 = gp.Trsf()
pos2.set_rotation(gp.Ax1(gp.Pnt(0,0,0), gp.Dir(0,0,1)), math.radians(45))
box2.move(pos2 * pos1)

sonde = brb.MakeSphere(r=0.5).solid()
pos3 = gp.Trsf()
pos3.set_rotation(gp.Ax1(gp.Pnt(0,0,0.), gp.Dir(0,0,1)), math.radians(20))
sonde.move(pos3 * pos1)

```

```python
from mod3d import BRepExtrema as bre

d1 = bre.DistShapeShape(sonde, corps_moteur)
d2 = bre.DistShapeShape(sonde, box1)
d3 = bre.DistShapeShape(sonde, box2)

print("Distance sonde - corps_moteur :", d1.value)
print("Distance sonde - box1  :", d2.value)
print("Distance sonde - box2  :", d3.value)
```

```python
print(f"Distance sonde - corps_moteur : {d1.value}, Points: {d1.point_on_shape1(1)}, {d1.point_on_shape2(1)}")
```

```python
from mod3d import StepControl as stc

writer = stc.Writer()
status = writer.transfer(corps_moteur, stc.AsIs)
status = writer.transfer(box1, stc.AsIs)
status = writer.transfer(box2, stc.AsIs)
status = writer.transfer(sonde, stc.AsIs)

write_status = writer.write("assembly.step")
```

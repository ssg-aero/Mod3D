```python
from mod3d import ShapeRenderer

shape_render = ShapeRenderer()
shape_render.angle_deflection = 5
```

```python
import mod3d

sphere = mod3d.BRepBuilderAPI.MakeSphere(5.0).shape()
box = mod3d.BRepBuilderAPI.MakeBox(10.0, 10.0, 10.0).shape()

# Move box to center on sphere
translation = mod3d.gp.Trsf()
translation.set_translation(mod3d.gp.Vec(-10.0, -10.0, -10.0))
box = box.moved(translation)

common = mod3d.BooleanOp.Common(sphere, box)
result = common.shape()

shape_render.add_shape(result, )

shape_render.render(background='lightgray')
```

```python
import mod3d

sphere = mod3d.BRepBuilderAPI.MakeSphere(5.0).shape()
box = mod3d.BRepBuilderAPI.MakeBox(10.0, 10.0, 10.0).shape()

# Move box to center on sphere
translation = mod3d.gp.Trsf()
translation.set_translation(mod3d.gp.Vec(-12.0, -12.0, -12.0))
box = box.moved(translation)

fuse = mod3d.BooleanOp.Fuse(box, sphere)
result = fuse.shape()

shape_render.clear()
shape_render.add_shape(result, )

shape_render.render(background='lightgray')
```

```python
# Create fillet
fillet_maker = mod3d.BRepFillet.MakeFillet(result)

# Add fillets to edges
for edge in fuse.section_edges():
    fillet_maker.add(1.0, edge)  # 3mm radius fillet

result = fillet_maker.shape()

shape_render.clear()
shape_render.add_shape(result, )

shape_render.render(background='lightgray')
```

```python
from mod3d import gp, Geom, BRepBuilderAPI, BRepOffsetAPI, TopoDS, GeomAbs
from pygbs import gbs as gbs

# use implict BSpline construction
crv1 = gbs.BSCurve3d(
    poles = [[0, 0, 0], [10, 0, 10], [20, 0, 0]],
    degree = 2,
    knots = [0,1],
    mults = [3,3]
)
e1 = BRepBuilderAPI.MakeEdge(crv1).edge()

crv2 = Geom.BSplineCurve(
    poles = [gp.Pnt(20, 0, 0), gp.Pnt(20,10,0)],
    degree = 1,
    knots = [0,1],
    multiplicities = [2,2]
)
e2 = BRepBuilderAPI.MakeEdge(crv2).edge()

crv3 = Geom.BSplineCurve(
    poles = [gp.Pnt(20,10,0), gp.Pnt(0,10,0)],
    degree = 1,
    knots = [0,1],
    multiplicities = [2,2]
)
e3 = BRepBuilderAPI.MakeEdge(crv3).edge()

crv4 = Geom.BSplineCurve(
    poles = [gp.Pnt(0,10,0), gp.Pnt(0,0,0)],
    degree = 1,
    knots = [0,1],
    multiplicities = [2,2]
)
e4 = BRepBuilderAPI.MakeEdge(crv4).edge()

shape_render.clear()
shape_render.add_shape([Geom.BSplineCurve(crv1), crv2, crv3, crv4], )

shape_render.render(background='lightgray')
```

```python
fill_op = BRepOffsetAPI.MakeFilling()

fill_op.add(e1, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e2, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e3, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e4, GeomAbs.Shape.C0, is_bound = True)

# fill_op.add(gp.Pnt(10,5,0))

fill_op.build()

assert fill_op.is_done(), "Filling operation failed"

face = fill_op.shape()

shape_render.clear()

shape_render.add_shape(face, )

shape_render.render(background='lightgray')
```

```python
fill_op = BRepOffsetAPI.MakeFilling() # Once Build is called, the object cannot be reused, so create a new one

fill_op.add(e1, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e2, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e3, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e4, GeomAbs.Shape.C0, is_bound = True)

fill_op.add(gp.Pnt(10,5,0))

fill_op.build()


assert fill_op.is_done(), "Filling operation failed"
face = fill_op.shape()

shape_render.clear()
shape_render.add_shape(face, )
shape_render.render(background='lightgray')
```

```python
from mod3d import BRepBuilderAPI, TopoDS

f1 = TopoDS.Face(BRepBuilderAPI.MakePrism(e1, gp.Vec(0,-10,0)).shape())
f3 = TopoDS.Face(BRepBuilderAPI.MakePrism(e3, gp.Vec(0,10,0)).shape())
            # py::arg("degree") = 3,
            # py::arg("nb_pts_on_cur") = 15,
            # py::arg("nb_iter") = 2,
            # py::arg("anisotropie") = false,
            # py::arg("tol_2d") = 0.00001,
            # py::arg("tol_3d") = 0.0001,
            # py::arg("tol_ang") = 0.01,
            # py::arg("tol_curv") = 0.1,
            # py::arg("max_deg") = 8,
            # py::arg("max_segments") = 9,
fill_op = BRepOffsetAPI.MakeFilling(
    degree = 5,
    nb_pts_on_cur = 15,
    nb_iter = 2,
    anisotropie = False,
    tol_2d = 0.00001,
    tol_3d = 0.0001,
    tol_ang = 0.01,
    tol_curv = 0.1,
    max_deg = 8,
    max_segments = 9,

) # Once Build is called, the object cannot be reused, so create a new one

fill_op.add(e1, f1,  GeomAbs.Shape.C1, is_bound = True)
fill_op.add(e2, GeomAbs.Shape.C0, is_bound = True)
fill_op.add(e3, f3, GeomAbs.Shape.C1, is_bound = True) # G2/C2, not supported
fill_op.add(e4, GeomAbs.Shape.C0, is_bound = True)

fill_op.build()

assert fill_op.is_done(), "Filling operation failed"
face = fill_op.shape()

shape_render.clear()
shape_render.add_shape(face, )
shape_render.add_shape(f3, )
shape_render.add_shape(f1, )
shape_render.render(background='lightgray')
```
